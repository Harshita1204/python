```python
import numpy as np;
import pandas as pd;
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data={
    'Area':[800,1000,1200,1400,1600,1800],
    'price':[40000,50000,60000,70000,80000,90000]
}
df=pd.DataFrame(data)
x=df[['Area']]
y=df[['price']]
model=LinearRegression()
model.fit(x,y)
new_area=[[2000]]
predicted_price=model.predict(new_area)
print("predicted price for 2000 sq,ft area",predicted_price[0])
plt.scatter(df['Area'],df['price'])
plt.plot(df['Area'],df['price'],color='red')
plt.xlabel('Area')
plt.ylabel('price')
plt.title('linear regression')
plt.show()
```
