```python
import numpy as np;
import pandas as pd;
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data={
    'trainingHours':[5,8,10,12,15,18],
    'productivityScore':[50,60,65,70,80,90]
}
df=pd.DataFrame(data)
x=df[['trainingHours']]
y=df['productivityScore']
model=LinearRegression()
model.fit(x,y)
new_trainingHours=[[7]]
new_trainingHours1=[[20]]
predicted_productivityScore=model.predict(new_trainingHours)
predicted_productivityScore1=model.predict(new_trainingHours1)

print("predicted productivity for 7 hours:  ",predicted_productivityScore)
print("predicted productivityfor 20 hours : ",predicted_productivityScore1)

plt.scatter(df['trainingHours'],df['productivityScore'])
plt.plot(df['trainingHours'],df['productivityScore'],color='pink')
plt.xlabel('trainingHours',color='purple')
plt.ylabel('productivityScore',color='green')
plt.title('Linear Regression',color='orange')
plt.show()
```
