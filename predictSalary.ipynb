```python
import numpy as np;
import pandas as pd;
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data={
    'experience':[1,2,3,4,5,6],
    'salary':[25000,30000,35000,40000,45000,50000]
}
df=pd.DataFrame(data)
x=df[['experience']]
y=df['salary']
model=LinearRegression()
model.fit(x,y)
new_experience=[[7]]
new_experience1=[[10]]
predicted_salary=model.predict(new_experience)
predicted_salary1=model.predict(new_experience1)

print("predicted salary for 7 years is: ",predicted_salary)
print("predicted salary for 10 years is : ",predicted_salary1)

plt.scatter(df['salary'],df['experience'])
plt.plot(df['salary'],df['experience'],color='pink')
plt.xlabel('salary')
plt.ylabel('experience')
plt.title('Linear Regression')
plt.show()
```
