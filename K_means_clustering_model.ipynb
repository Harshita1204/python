```python
from sklearn.cluster import KMeans
import pandas as pd

data = pd.DataFrame({
    'age':[21,23,45,48,22,46],
    'income':[20000,25000,80000,85000,22000,82000]
})
model= KMeans(n_clusters=2)
model.fit(data)
#print(model.predict(data))
clusters=model.labels_
print(clusters)
```
